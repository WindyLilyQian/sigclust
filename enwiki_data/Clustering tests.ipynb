# 1st the basic sigclust


```python
p_value, clusters = sigclust.sigclust.sigclust(features)
```

**Explanation of output:**

(1 - Cluster index) ~ the proportion of deviation explained by the clusters (as opposed one single gaussian)
Median absolute deviation(MAD) is a measure of variance (Standard Deviation)
Variance for the background noise = (MAD * scaling_factor)^2

**Then we simulate**

Assume that there is one cluster (a multivariate gaussian).  Simulate draws from potential single clusters with these properties.  If our empirical cluster index is very different from the distribution of simulated cluster indexes then, our clusters explain the data better than the null hypothesis (single cluster).
